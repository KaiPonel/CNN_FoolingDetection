```python
import matplotlib.pyplot as plt

timestamps = []
memory_usage = []
with open('m_leaks2.dat', 'r') as f:
    for line in f:
        if line.startswith('MEM'):
            parts = line.split()
            memory_usage.append(float(parts[1]))
            timestamps.append(float(parts[2]))

plt.plot(timestamps, memory_usage)
plt.xlabel('Time (seconds)')
plt.ylabel('Memory usage (MiB)')
plt.show()
```

```python
timestamps = []
memory_usage = []
with open('m_leaks2.dat', 'r') as f:
    for line in f:
        if line.startswith('MEM'):
            parts = line.split()
            memory_usage.append(float(parts[1]))
            timestamps.append(float(parts[2]))

timestamps = [(t - timestamps[0]) / 10 for t in timestamps]


plt.plot(timestamps, memory_usage)
plt.xlabel('Time (seconds)')
plt.ylabel('Memory usage (MiB)')
plt.show()
```
